```python
#Matplotlib課題004
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(2,3,tight_layout=True)#subplots()、複数のプロットを一つにまとめる

#折れ線グラフ
x = [0,1,2,3] #X軸用意
y = [1,2,-5,2] #Y軸用意
ax[0,0].plot(x,y) #折れ線グラフをプロット

#sin関数グラフ
x = np.linspace(0, 10, 100) #X軸、0から10までの幅で配列を等間隔に100個用意
y = 2 + 2 * np.sin(2 * x) #Y軸用意
ax[0,1].plot(x,y) #sin関数グラフをプロット

#ヒストグラム
x = np.random.normal(15,5,2000) #平均が15, 標準偏差5の標準正規分布から2000個の乱数を生成
plt.xlim(-10,40) #左右対称を強調するためX軸を調整
ax[0,2].hist(x,bins=30) #ヒストグラムをプロット、histメソッドの棒(bins)の数はデフォルトは10個なので30個に調整

#散布図
np.random.seed(3) #データを用意
x = 4 + np.random.normal(0,2,60) #X軸用意
y = 4 + np.random.normal(0,2,len(x)) #Y軸用意
ax[1,0].scatter(x,y) #散布図をプロット

plt.show()

```
